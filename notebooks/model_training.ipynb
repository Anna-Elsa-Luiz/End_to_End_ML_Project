### Dropped the dependant or target variable .Here **price** 

```python
X = data.drop(labels=['price'],axis=1)
X
```

### Segregating into categorical and numerical 

```python
categorical = X.select_dtypes(include='object').columns
categorical
```

```python
numerical = X.select_dtypes(exclude='object').columns
numerical
```

### Define the custom ranking for each ordinal variable

```python
cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']

```

### Importing necessary libraries 

```python
from sklearn.impute import SimpleImputer ## HAndling Missing Values
from sklearn.preprocessing import StandardScaler # HAndling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

In above steps


could keep all the methods in a sequence

### Creating the the numerical and categorical pipelines

Automating  every steps 

```python
num_pipeline = Pipeline(
              
              steps = [
                       ('imputer',SimpleImputer()),
                       ('scaler', StandardScaler())
              ]

)
```

```python
cat_pipeline = Pipeline(


                steps=[
                       ('imputer',SimpleImputer(strategy='most_frequent')),
                       ('OrdinalEncoder',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories]))
                ]
)
```

```python
preprocessor = ColumnTransformer(

              [
                 ('numerical_pipeline',num_pipeline,numerical),
                 ('categorical_pipeline',cat_pipeline,categorical)
              ]

)
```

### Splitting the dataset into Train and Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.transform(X_test)
```
